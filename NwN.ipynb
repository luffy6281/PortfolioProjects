```python
# convert "date" column to index 
#index_col ="date", makes "date" column
df = pd.read_csv("SampleA.csv", parse_dates =["Post_creation_date"], index_col ="Post_creation_date")

df.head()
```

```python
#Drop columns not for use here
df.drop(df.columns[[1,2,3]], axis = 1, inplace = True)
df
```

```python
# settings to display all columns
pd.set_option("display.max_columns", None)

#To display column with page FB003 only
dfA = df.loc[df.Page_ID=='FB003', :]
dfA
```

```python
# Resampling the data on weekly basis for Page FB003
#Calculating Mean value of all fields over a Week

dfA_M = dfA.resample('W').mean()                    ##Mean value of different fields over a week for FB003 page

dfA_M

#Engagement is the sum of likes, share and comments on a post
#Mean engagement with post in a week
dfA_M['Mean_Engagement'] = dfA_M.apply(lambda row: row.Post_reactions + row.Post_comments + row.Post_shares, axis = 1) 

#Average Percent of people who engaged with the post after they saw the post (over a week)
dfA_M["Mean_Percent_engagement"] = ((dfA_M["Mean_Engagement"] / dfA_M["Post_reach"])*100)

#Mean Ratio of complete video views to the video views (over a week)
dfA_M["MeanRatio_completion"] = (dfA_M["Completed_video_views"] / dfA_M["Video_views"])
dfA_M.head(467)
```

```python
# Resampling the data on weekly basis for Page FB003
#Calculating Total value of all fields over a Week

dfA_T = dfA.resample('W').sum()          ##Total value of different fields over a week for FB003 page
dfA_T

#Engagement is the sum of likes, share and comments on a post
#Total engagement with the posts in a week 
dfA_T['Sum_Engagement'] = dfA_T.apply(lambda row: row.Post_reactions + row.Post_comments + row.Post_shares, axis = 1) 

#Total Percent of people who engaged with the post after they saw the post (over a week)
dfA_T["Sum_Percent_engagement"] = ((dfA_T["Sum_Engagement"] / dfA_T["Post_reach"])*100)

#Total Ratio of complete video views to video views (over a week)
dfA_T["Sum_Ratio_completion"] = (dfA_T["Completed_video_views"] / dfA_T["Video_views"])
dfA_T
```
