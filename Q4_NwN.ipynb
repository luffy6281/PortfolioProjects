```python
import pandas as pd
import numpy as np
import datetime
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
import io
df = pd.read_csv(io.BytesIO(uploaded['SampleData.csv']))
```

```python
df.head()
```

```python
#Drop the columns that will not be in use
df.drop(df.columns[[2]], axis = 1, inplace = True)
df
```

```python
df[['Post_creation_date']] = df[['Post_creation_date']].apply(pd.to_datetime)
```

```python
# settings to display all columns
pd.set_option("display.max_columns", None)

#To display column with page FB003 only
dfA = df.loc[df.Page_ID=='FB003', :]
dfA
```

```python
#Selecting only video posts for FB003
dfAa= dfA.dropna()
dfAa
```

```python

pd.set_option('display.max_rows', None)
```

```python
#Engagement is the sum of likes, share and comments on a video post   ###FB003
#Mean engagement with post in a week
dfAa['Daily_Engagement'] = dfAa.apply(lambda row: row.Post_reactions + row.Post_comments + row.Post_shares, axis = 1) 

#Engagement per second of the video post (over a week)
dfAa["Engagement/sec"] = ((dfAa["Daily_Engagement"] / dfAa["Video_length (s)"]))

#Average Percent of people who engaged with the post after they saw the post (over a week)
dfAa["Daily_Percent_engagement"] = ((dfAa["Daily_Engagement"] / dfAa["Post_reach"])*100)

#Mean Ratio of complete video views to the video views (over a week)
dfAa["DailyRatio_completion"] = (dfAa["Completed_video_views"] / dfAa["Video_views"])
dfAa
```
